```python
# Persiapan membuat data
import pandas as pd
import random

pelajaran = ['Matematika', 'Fisika', 'Biologi', 'Kimia', 'Geologi', 'Ekonomi']
kelas = ['12A', '12B', '12C', '12D']

hasil_nilai = {
    'siswa_id' : range(1, 21),
    'pelajaran' : [random.choice(pelajaran) for i in range(20)],
    'nilai' : [random.choice(range(0, 100)) for i in range(20)],
    'jenis-kelamin' : [random.choice(['laki-laki', 'perempuan']) for i in range(20)],
    'kelas' : [random.choice(kelas) for i in range(20)]
}

df = pd.DataFrame(hasil_nilai)

df
```

```python
#contoh indexing
clm1=['nilai','kelas']
ni_ke= df.loc[:,clm1]
ni_ke
```

```python
clm2=['pelajaran','nilai']
pe_ni= df.loc[4:9,clm2]
pe_ni
```

```python
#Filter DataFrame Dengan Boolean
pel1=df.pelajaran =='Ekonomi'
pel2=df.pelajaran =='Fisika'
df[(pel1) |(pel2)]
```

```python
#Filter DataFrame Dengan Boolean
nil=df.nilai>50
pel3=df.pelajaran=='Kimia'
df[(nil)&(pel3)]
```

```python
nil2=df.nilai>70
jk=df['jenis-kelamin']=='perempuan'
pel4=df.kelas=='12C'
df[(nil2) & (jk)&(pel4)]
```
